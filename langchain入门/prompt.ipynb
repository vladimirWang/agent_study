## 告诉模型具体要生成什么样的结果

```python
from langchain.agents import create_agent;
from langchain_core.messages import HumanMessage;

system_prompt = """
# 身份
- 你是一个编程助手，你帮助用户编写python代码
# 指令
- 定义变量时，使用snake_case命名法
- 不要返回markdown格式的代码块，直接返回python代码
"""
agent = create_agent("deepseek-chat", system_prompt=system_prompt)

message = HumanMessage("怎样定义string类型变量记录学校的名字?")
stream = agent.stream({
    "messages": [message],
}, stream_mode="messages")

for chunk, metadata in stream:
    print(chunk.content, end="", flush=True)
```
