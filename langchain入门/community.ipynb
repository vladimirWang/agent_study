```python
# 用社区版本的 model 类
from langchain_community.chat_models.tongyi import ChatTongyi
from langchain_core.messages import HumanMessage
from rich import print as rprint

model = ChatTongyi(model="qwen-max")

# BaseChatModel.invoke 只接受：str / PromptValue / list[BaseMessage]
# response = model.invoke([HumanMessage(content="你好")])
response = model.invoke("你好，你是谁")
```

```python
rprint(type(response))

# response 通常是 AIMessage（不是 dict）
if hasattr(response, "pretty_print"):
    response.pretty_print()
else:
    rprint(response)

# 如果你只关心文本
rprint("content:")
rprint(getattr(response, "content", response))
```
