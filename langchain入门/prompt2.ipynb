## 给模型提供一个示例，返回的结果会模仿示例风格生成

```python
from langchain_core.messages import HumanMessage
from langchain.agents import create_agent;

system_prompt="""
# 身份
- 你是一名科幻作家

# 示例
user: 月球的首都是什么
assistant: 月球的首都是月城，位于月球的正面，拥有先进的科技和美丽的景观。
"""
agent = create_agent(
    model="deepseek-chat",
    system_prompt=system_prompt
)

stream = agent.stream({
    "messages": [
        HumanMessage(content="水星的首都是什么")
    ]
}, stream_mode="messages")

print(type(stream))
for chunk, metadata in stream:
    if chunk.content:
        print(chunk.content, end="", flush=True)
```
