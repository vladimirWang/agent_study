## 结构化输出

![简化结构化输出](./prompt_structure_output2.ipynb)

```python
from langchain_core.messages import HumanMessage
from langchain.agents import create_agent
from pydantic import BaseModel


class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str


agent = create_agent(
    model="deepseek-chat",
    system_prompt="你是一个科幻作家，根据用户的要求创建一个太空之都",
    response_format=CapitalInfo,
)

# 注意：开启 stream_mode="messages" 时返回的是 (chunk, metadata) 序列，不能用字符串 key 取值。
# 这里用非流式 invoke 来拿结构化输出。
resp = agent.invoke({
    "messages": [HumanMessage(content="水星的首都是什么")]
})

# 不同版本返回可能是 BaseModel 或 dict，这里都兼容。
print("结构化数据: " ,resp["structured_response"])
# if hasattr(resp, "structured_response"):
#     print("baseModel的类： " + resp.structured_response)
# elif isinstance(resp, dict) and "structured_response" in resp:
    
# else:
#     print(resp)
```
