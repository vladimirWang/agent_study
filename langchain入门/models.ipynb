```python
from langchain.agents import create_agent

agent = create_agent("deepseek-chat")

print(agent, type(agent))

messages = agent.stream({"messages": [{"role": "user", "content": "你好，你是谁？"}]}, stream_mode="messages")

# print(type(messages), messages)
for chunk, metadata in messages:
    if chunk.content:
        print(chunk.content, end="", flush=True)
    else:
        print("没有content")
```
