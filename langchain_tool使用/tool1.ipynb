```python
from langchain.tools import tool
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage
from pydantic import BaseModel, Field
from typing import Literal

class WeatherInput(BaseModel):
    location: str = Field(description="地点名称")
    units: Literal["celsius", "fahrenheit"] = Field(default="celsius", description="温度单位")
    include_forecast: bool = Field(default=False, description="是否包含预报信息")

@tool
def get_weather(location: str) -> str:
    """获取指定地点的天气信息"""
    return f"the weather in {location} is sunny"


@tool(args_schema=WeatherInput)
def get_weather_structured(location: str, units: Literal["celsius", "fahrenheit"] = "celsius", include_forecast: bool = False) -> str:
    """获取指定地点的天气信息"""
    result = f"the weather in {location} is sunny"
    if units == "celsius":
        result += f" in {28} celsius"
    else:
        result += f" in {68} fahrenheit"
    if include_forecast:
        result += f" and is rainy in next 3 days"
    return result

agent = create_agent(
    model="deepseek-chat",
    tools=[get_weather_structured],
)

resp = agent.invoke({
    "messages": [HumanMessage(content="请告诉我北京的天气")]
})

for message in resp["messages"]:
    print(message.content)
```
