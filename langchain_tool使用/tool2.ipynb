```python
from openai import OpenAI
from dotenv import load_dotenv
import os
from langchain.tools import tool
print("1")
api_key=os.getenv("DEEPSEEK_API_KEY")

@tool
def square_root(n: float) -> float:
    """计算一个数的平方根"""
    return n**0.5;

@tool
def getWeather(location: str) -> str:
    """Get the current weather in a given location"""
    return f"The current weather in {location} is 75 degrees and sunny."
```

```python
from pydantic import BaseModel, Field
from typing import Literal


class WeatherInput(BaseModel):
    location: str = Field(description="the location to get the weather for")
    units: Literal["celsius", "fahrenheit"] = Field(
        default="celsius", description="the units to return the weather in"
    )
    include_forecast: bool = Field(
        default=False, description="include 5 days forecast in the response"
    )


@tool(args_schema=WeatherInput)
def get_weather_structured(
    location: str, units: str = "celsius", include_forecast: bool = False
) -> str:
    """get current weather for a location, with optional forecast"""
    temp = 22 if units == "celsius" else 72
    result = f"The current weather in {location} is {temp} degrees {units}."
    if include_forecast:
        result += " The forecast for the next few days is sunny with a chance of rain."
    return result
```

```python
from langchain.agents import create_agent

agent = create_agent(
    "deepseek-chat",
    tools=[get_weather_structured, square_root]
)
```
