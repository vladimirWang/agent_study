```python
from langchain.agents import create_agent
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
from pydantic import BaseModel, Field
from typing import List
from langchain_tavily import TavilySearch

class Reference(BaseModel):
    title: str = Field(..., description="参考来源的标题")
    url: str = Field(..., description="参考来源的URL链接")

class AnswerInfo(BaseModel):
    answer: str = Field(..., description="答案的文本内容")
    references: List[Reference] = Field(..., description="答案的参考来源列表")

tavily = TavilySearch(max_results=3)
@tool
def web_search(query: str):
    """使用Tavily进行网络搜索，并返回答案和参考来源"""
    return tavily.invoke(query)

agent = create_agent(
    "deepseek-chat",
    tools=[web_search],
    response_format=AnswerInfo,
    system_prompt="你是一个有用的助手，可以使用工具来帮助用户获取信息。"
)
resp = agent.invoke({
    "messages": [HumanMessage(content="什么是蒸蚌？")]
})

# for message in resp['messages']:
#     print(message.pretty_print())

print(resp['structured_response'])
```
